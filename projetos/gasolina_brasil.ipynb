### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
gasolina['PREÇO MÉDIO REVENDA'] = gasolina['PREÇO MÉDIO REVENDA'].astype(float)
gasolina[gasolina['ano_mes'] == '2008-08']['PREÇO MÉDIO REVENDA'].mean()
#df.info()
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
gasolina[(gasolina['ano_mes'] == '2014-05') & (gasolina['ESTADO'] == 'SAO PAULO')]['PREÇO MÉDIO REVENDA'].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
df_preco = gasolina[['ESTADO','PREÇO MÉDIO REVENDA','DATA FINAL']]
df_preco[df_preco['PREÇO MÉDIO REVENDA'] > 5.0].head(20)
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_aux = gasolina[(gasolina["DATA FINAL"].apply(lambda x: x.year) == 2012)]
df_aux[df_aux["REGIÃO"] == "SUL"]["PREÇO MÉDIO REVENDA"].mean()
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
gasolina["MES"] = gasolina["DATA FINAL"].apply(lambda x: x.month)
df_rio = gasolina[gasolina["ESTADO"] == "RIO DE JANEIRO"]
df_month_rio = df_rio.groupby("ano_mes")[["PREÇO MÉDIO REVENDA", "MES"]].last()
((df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100).rolling(3).sum()
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_max = gasolina.groupby("ano_mes").max()["PREÇO MÉDIO REVENDA"].sort_index()
idx_max = gasolina.groupby('ano_mes')["PREÇO MÉDIO REVENDA"].idxmax()
df_min = gasolina.groupby("ano_mes").min()["PREÇO MÉDIO REVENDA"].sort_index()
idx_min = gasolina.groupby('ano_mes')["PREÇO MÉDIO REVENDA"].idxmin()
df_diff = pd.DataFrame()
```

```python
df_diff["abs_diff"] = df_max - df_min
df_diff["percent_diff"] = (df_max/ df_min) / df_min * 100
df_diff["max"] = df_max
df_diff["min"] = df_min
```
